```
#!fsharp
open Deedle
open Plotly.NET
```

```
#!fsharp
let weather = Frame.ReadCsv "/home/gregor/source/repos/FSharpForDataScience/datasets/nycflights13/weather.csv"
```

```
#!fsharp
weather
```

```
#!fsharp
let weatherWithMonth =
    weather
    |> Frame.addCol "Month" (weather.GetColumn<DateTime> "time_hour" |> Series.map (fun _ value -> value.Month))
    |> Frame.sliceCols [ "origin"; "Month"; "temp" ]

weatherWithMonth
```

```
#!fsharp
let charts =
    weatherWithMonth
    |> Frame.groupRowsByInt "Month"
    |> Frame.nest
    |> Series.dropMissing
    |> Series.values
    |> Seq.mapi (fun index frame ->
        frame?temp
        |> Series.dropMissing
        |> Series.values
        |> fun data -> Chart.Histogram(data, nBinsx = 30, Name = (index + 1 |> sprintf "Month %i"))
        |> Chart.withLayout (Layout.init(Bargap = 0.1)))
    |> Seq.splitInto 4
    |> Seq.rev

Chart.Grid(charts)
|> Chart.withLegend false
|> Chart.withTitle "Temperatures (Jannuary - December)"
|> Chart.withX_AxisStyle(title = "Temperature (F)")
|> Chart.withY_AxisStyle(title = "Frequency")
|> Chart.withSize (900., 600.)
```
